# Importing Libraries and configing plot styles

```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
matplotlib.rcParams['figure.figsize'] = (12,8)
import numpy as np

pd.options.mode.chained_assignment = None
```

# Loading and displaying data

```python
df = pd.read_csv('./movies.csv')
df.head()
```

# Checking if there is any missing data

```python
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

# Removing the null values

```python
df = df.dropna()
```

# Data Types for our columns

```python
df.dtypes
```

# Change datatype of columns

```python
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

# Create correct year column

```python
df['yearcorrect'] = df['released'].str.extract(r'(\d{4})')
df
```

# Sorting Values by Gross

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

# Drop any Duplicates

```python
df.drop_duplicates()
```

```python
# Budget high correlation
# Company high correlation
```

# Scatter plot with budget vs gross

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show
```

```python
df.head()
```

# Plot budget vs gross using seaborn

```python
sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'red'}, line_kws={'color': 'blue'})
```

# Looking at Correlations

```python
# Select only the numeric columns
df_numeric = df.select_dtypes(include='number')
# Compute the correlation matrix
df_numeric.corr() #pearson

# High correlation between budget and gross was Right!
```

```python
df_numeric.corr(method='kendall') #kendall
```

```python
df_numeric.corr(method='spearman') #spearman
```

```python
correlation_matrix = df_numeric.corr() #pearson

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for numeric correlations')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
